```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np


def plot_threat_level(threat_data, threat_name):
    threat_factors = threat_data['Threat Factors']
    threat_levels = threat_data['Current EV Market Position']

    y_pos = np.arange(len(threat_factors))
    bar_width = 0.35

    fig, ax = plt.subplots()
    bars = ax.barh(y_pos, threat_levels, bar_width, align='center')

    ax.set_yticks(y_pos)
    ax.set_yticklabels(threat_factors)
    ax.invert_yaxis()  # labels read from top to bottom
    ax.set_xlabel('Threat Level (Low to High)')
    ax.set_title(f'{threat_name} Threat Level Analysis')

    for bar in bars:
        width = bar.get_width()
        ax.text(width, bar.get_y() + bar.get_height() / 2, f'{width}', ha='left', va='center')

    plt.show()


# 假设数据已经读取到一个DataFrame中
data = {
    'Competitive Rivalry': {
        'Threat Factors': ['Number of Competitors', 'Industry Growth Rate', 'Product Differentiation',
                           'Price Wars', 'Fixed and Storage Costs', 'Exit Barriers'],
        'Current EV Market Position': [5, 3, 3, 5, 5, 5]
    },
    'Threat of New Entrants': {
        'Threat Factors': ['Capital Requirements', 'Supply-Side Economies of Scale', 'Demand-Side Network Effects',
                           'Government Regulations', 'Incumbent Response to New Entrants', 'Access to Distribution Channels'],
        'Current EV Market Position': [5, 4, 3, 5, 5, 5]
    },
    'Threat of Substitutes': {
        'Threat Factors': ['Substitute Products’ Performance', 'Price of Substitutes', 'Customer Switching Costs',
                           'Consumer Behavior', 'Technological Disruption'],
        'Current EV Market Position': [3, 3, 3, 3, 4]
    },
    'Bargaining Power of Suppliers': {
        'Threat Factors': ['Number of Suppliers', 'Supplier Concentration', 'Switching Costs for Automakers',
                           'Threat of Supplier Forward Integration', 'Availability of Substitute Materials'],
        'Current EV Market Position': [3, 3, 2, 2, 3]
    },
    'Bargaining Power of Buyers': {
        'Threat Factors': ['Number of Buyers', 'Buyer Information Availability', 'Switching Costs for Buyers',
                           'Price Sensitivity', 'Fleet Buyers’ Influence'],
        'Current EV Market Position': [2, 5, 3, 5, 5]
    }
}

for threat_name, threat_data in data.items():
    df = pd.DataFrame(threat_data)
    plot_threat_level(df, threat_name)
```
